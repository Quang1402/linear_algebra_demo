```python
import numpy as np

np.set_printoptions(precision=4, suppress=True)

def print_header(title):
    print("\n" + "="*80)
    print(title)
    print("="*80)
```

```python
print_header("Question 1: (A^T)^T = A")

A = np.random.randn(3, 4)
lhs = (A.T).T
rhs = A

print("A:\n", A)
print("\n(A.T).T:\n", lhs)
print("\nTất cả phần tử bằng nhau? ->", np.allclose(lhs, rhs))
```

```python
print_header("Question 2: A^T + B^T = (A + B)^T")

A = np.random.randn(3, 4)
B = np.random.randn(3, 4)

lhs = A.T + B.T
rhs = (A + B).T

print("A:\n", A)
print("\nB:\n", B)
print("\nA.T + B.T:\n", lhs)
print("\n(A + B).T:\n", rhs)
print("\nTất cả phần tử bằng nhau? ->", np.allclose(lhs, rhs))
```

```python
print_header("Question 3: A + A^T có luôn đối xứng không?")

A = np.random.randn(4, 4)

S = A + A.T
S_T = S.T

print("A:\n", A)
print("\nS = A + A.T:\n", S)
print("\nS.T:\n", S_T)
print("\nS có bằng S.T? ->", np.allclose(S, S_T))
```

```python
print_header("Question 4: len(X) với X shape (2, 3, 4)")

X = np.random.randn(2, 3, 4)
print("X shape:", X.shape)
print("len(X):", len(X))
```

```python
print_header("Question 5: len(X) cho tensor arbitrary shape")

shapes = [(5, 7),
          (2, 3, 4),
          (10, 1, 3, 8)]

for shp in shapes:
    X = np.zeros(shp)
    print("Shape:", X.shape, "-> len(X) =", len(X),
          "(bằng X.shape[0] =", X.shape[0], ")")
```

```python
print_header("Question 6: A / A.sum(axis=1)")

A = np.random.rand(3, 4)
row_sum = A.sum(axis=1)

print("A:\n", A)
print("\nA.sum(axis=1):", row_sum, " | shape:", row_sum.shape)

B = A / row_sum[:, None]
print("\nB = A / A.sum(axis=1) (dùng broadcast):\n", B)

print("\nSum theo hàng của B (axis=1):", B.sum(axis=1))
```

```python
print_header("Question 7: Manhattan distance")

def manhattan_distance(p, q):
    p = np.array(p)
    q = np.array(q)
    return np.sum(np.abs(p - q))

p = np.array([1, 2])
q = np.array([5, 7])

print("Điểm p:", p, " | điểm q:", q)
print("Khoảng cách Manhattan (|Δavenue| + |Δstreet|) =", manhattan_distance(p, q))
print("Nếu được đi chéo (Euclid) thì khoảng cách =", np.linalg.norm(q - p))
print("\nTrong Manhattan, ta không được đi chéo, nên dùng norm-1.")
```

```python
print_header("Question 8: Summation shapes với tensor (2, 3, 4)")

X = np.arange(2*3*4).reshape(2, 3, 4)
print("X shape:", X.shape)
print("X:\n", X)

sum0 = X.sum(axis=0)
sum1 = X.sum(axis=1)
sum2 = X.sum(axis=2)

print("\nSum axis=0 -> shape", sum0.shape, ":\n", sum0)
print("\nSum axis=1 -> shape", sum1.shape, ":\n", sum1)
print("\nSum axis=2 -> shape", sum2.shape, ":\n", sum2)
```

```python
print_header("Question 9: linalg.norm cho tensor nhiều trục")

X = np.random.randn(2, 3, 4)
norm_builtin = np.linalg.norm(X)

sum_squares = np.sum(X**2)
norm_manual = np.sqrt(sum_squares)

print("X shape:", X.shape)
print("\nnp.linalg.norm(X) =", norm_builtin)
print("sqrt(sum(X**2)) =", norm_manual)
print("\nHai giá trị gần như bằng nhau? ->", np.allclose(norm_builtin, norm_manual))
```

```python
print_header("Question 10: Cost (AB)C vs A(BC) với kích thước lớn")

def cost(m, n, p):
    return m * n * p

a1, a2 = 10, 16
b2, b3 = 5, 5
c3, c4 = 5, 14

cost_AB = cost(2**10, 2**16, 2**5)
cost_ABC_1 = cost_AB + cost(2**10, 2**5, 2**14)

cost_BC = cost(2**16, 2**5, 2**14)
cost_ABC_2 = cost_BC + cost(2**10, 2**16, 2**14)

print("Cost (AB)C ≈ 2^31 + 2^29 ~", cost_ABC_1)
print("Cost A(BC) ≈ 2^35 + 2^40 ~", cost_ABC_2)
print("Tỷ lệ (A(BC)) / ((AB)C) ~", cost_ABC_2 / cost_ABC_1)

print("\n(AB) có shape 2^10 x 2^5 (2^15 phần tử) => ma trận trung gian khá nhỏ.")
print("BC có shape 2^16 x 2^14 (2^30 phần tử) => ma trận trung gian khổng lồ.")
```

```python
print_header("Question 11: AB vs AC^T và C = B^T")

m, n, p = 64, 128, 32
A = np.random.randn(m, n)
B = np.random.randn(n, p)
C = np.random.randn(p, n)

AB = A @ B
ACt = A @ C.T

print("A shape:", A.shape)
print("B shape:", B.shape)
print("C shape:", C.shape)
print("AB shape:", AB.shape)
print("A @ C.T shape:", ACt.shape)
print("\nCost lý thuyết 2 trường hợp giống nhau vì đều là m*n*p phép toán.")

C_view = B.T
ACt_view = A @ C_view.T

print("\nTrường hợp đặc biệt C = B^T:")
print("C_view shape (B.T):", C_view.shape)
print("C_view.T shape:", C_view.T.shape)
print("A @ C_view.T shape:", ACt_view.shape)
print("A @ C_view.T có bằng AB? ->", np.allclose(ACt_view, AB))
```

```python
print_header("Question 12: Stack [A,B,C] và slice lại B")

A = np.random.randn(100, 200)
B = np.random.randn(100, 200)
C = np.random.randn(100, 200)

T = np.stack([A, B, C], axis=2)

print("A, B, C shape:", A.shape)
print("Tensor T shape:", T.shape)

B_recovered = T[:, :, 1]

print("\nB_recovered shape:", B_recovered.shape)
print("B_recovered có bằng B gốc không? ->", np.allclose(B_recovered, B))
```
