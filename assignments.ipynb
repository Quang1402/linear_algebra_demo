```python
import numpy as np

np.set_printoptions(precision=4, suppress=True)

def print_header(title):
    print("\n" + "="*80)
    print(title)
    print("="*80)
```

```python
print_header("Question 1: (A^T)^T = A")

A = np.random.randn(3, 4)
lhs = (A.T).T
rhs = A

print("A:\n", A)
print("\n(A.T).T:\n", lhs)
print("\nTất cả phần tử bằng nhau? ->", np.allclose(lhs, rhs))
```

```python
print_header("Question 2: A^T + B^T = (A + B)^T")

A = np.random.randn(3, 4)
B = np.random.randn(3, 4)

lhs = A.T + B.T
rhs = (A + B).T

print("A:\n", A)
print("\nB:\n", B)
print("\nA.T + B.T:\n", lhs)
print("\n(A + B).T:\n", rhs)
print("\nTất cả phần tử bằng nhau? ->", np.allclose(lhs, rhs))
```

```python
print_header("Question 7: Manhattan distance")

def manhattan_distance(p, q):
    p = np.array(p)
    q = np.array(q)
    return np.sum(np.abs(p - q))

p = np.array([1, 2])
q = np.array([5, 7])

print("Điểm p:", p, " | điểm q:", q)
print("Khoảng cách Manhattan (|Δavenue| + |Δstreet|) =", manhattan_distance(p, q))
print("Nếu được đi chéo (Euclid) thì khoảng cách =", np.linalg.norm(q - p))
print("\nTrong Manhattan, ta không được đi chéo, nên dùng norm-1.")
```

```python
print_header("Question 10: Cost (AB)C vs A(BC) với kích thước lớn")

def cost(m, n, p):
    return m * n * p

a1, a2 = 10, 16
b2, b3 = 5, 5
c3, c4 = 5, 14

cost_AB = cost(2**10, 2**16, 2**5)
cost_ABC_1 = cost_AB + cost(2**10, 2**5, 2**14)

cost_BC = cost(2**16, 2**5, 2**14)
cost_ABC_2 = cost_BC + cost(2**10, 2**16, 2**14)

print("Cost (AB)C ≈ 2^31 + 2^29 ~", cost_ABC_1)
print("Cost A(BC) ≈ 2^35 + 2^40 ~", cost_ABC_2)
print("Tỷ lệ (A(BC)) / ((AB)C) ~", cost_ABC_2 / cost_ABC_1)

print("\n(AB) có shape 2^10 x 2^5 (2^15 phần tử) => ma trận trung gian khá nhỏ.")
print("BC có shape 2^16 x 2^14 (2^30 phần tử) => ma trận trung gian khổng lồ.")
```

```python
print_header("Question 20: Eigenvalues & eigenvectors of A = [[2,1],[1,2]]")

A = np.array([[2, 1],
              [1, 2]], dtype=float)

# Symmetric matrix => use eigh (real eigenvalues, orthonormal eigenvectors)
eigenvalues, eigenvectors = np.linalg.eigh(A)  # ascending order

print("A =\n", A)
print("\nEigenvalues (ascending) =", eigenvalues)
print("Eigenvectors (columns correspond to eigenvalues) =\n", eigenvectors)

print("\nVerification: residuals A v - lambda v")
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]
    residual = A @ v - lam * v
    print(f"i={i}, lambda={lam:.4f}, residual={residual}")

# Short explanation (printed)
print("\nExplanation:")
print("A is symmetric, so np.linalg.eigh returns real eigenvalues and orthonormal eigenvectors.")
print("Each column v satisfies A v = lambda v (residuals above are ~ 0).")
print("The eigenvalues should be close to [1, 3] (ascending).")
```

```python
print_header("Question 13: Angle between v1 and v2")

v1 = np.array([1, 0, -1, 2], dtype=float)
v2 = np.array([3, 1, 0, 1], dtype=float)

dot = float(v1 @ v2)
n1 = float(np.linalg.norm(v1))
n2 = float(np.linalg.norm(v2))

cos_theta = dot / (n1 * n2)
cos_theta_clipped = float(np.clip(cos_theta, -1.0, 1.0))

theta_rad = float(np.arccos(cos_theta_clipped))
theta_deg = float(np.degrees(theta_rad))

print("v1 =", v1)
print("v2 =", v2)
print("\nv1 · v2 =", dot)
print("||v1|| =", n1)
print("||v2|| =", n2)
print("cos(theta) =", cos_theta)
print("theta (radians) =", theta_rad)
print("theta (degrees) =", theta_deg)

print("\nExplanation:")
print("Angle is computed by cos(theta) = (v1·v2) / (||v1|| ||v2||), then theta = arccos(cos(theta)).")
print("Clipping is used only to avoid tiny numerical errors pushing cos(theta) slightly outside [-1, 1].")
```

```python
print_header("Question 14: True/False: [[1,2],[0,1]] and [[1,-2],[0,1]] are inverses")

A = np.array([[1,  2],
              [0,  1]], dtype=float)

B = np.array([[1, -2],
              [0,  1]], dtype=float)

I = np.eye(2)

AB = A @ B
BA = B @ A

print("A =\n", A)
print("\nB =\n", B)
print("\nA @ B =\n", AB)
print("\nB @ A =\n", BA)

print("\nIs A @ B equal to I? =", np.allclose(AB, I))
print("Is B @ A equal to I? =", np.allclose(BA, I))

print("\nExplanation:")
print("Two matrices are inverses if both A B = I and B A = I.")
print("The products printed above show whether the inverse relationship holds in both directions.")
```

```python
print_header("Question 15: Area scaling after transforming by A = [[2,3],[1,2]]")

area_original = 100.0  # m^2
A = np.array([[2, 3],
              [1, 2]], dtype=float)

detA = float(np.linalg.det(A))
scale = abs(detA)
area_new = area_original * scale

print("A =\n", A)
print("\ndet(A) =", detA)
print("Area scale factor = |det(A)| =", scale)
print("Original area =", area_original)
print("New area =", area_new)

print("\nExplanation:")
print("A linear transform in 2D scales area by |det(A)|.")
print("So the transformed area equals original_area * |det(A)|.")
```

```python
print_header("Question 16: Which sets of vectors are linearly independent?")

def is_linearly_independent(vectors, tol=1e-12):
    M = np.column_stack(vectors).astype(float)
    r = np.linalg.matrix_rank(M, tol=tol)
    return r == M.shape[1], r, M

sets = [
    (
        "Set 1",
        [np.array([1, 0, -1], dtype=float),
         np.array([2, 1, -1], dtype=float),
         np.array([3, 1,  1], dtype=float)]
    ),
    (
        "Set 2",
        [np.array([3, 1, 1], dtype=float),
         np.array([1, 1, 1], dtype=float),
         np.array([0, 0, 0], dtype=float)]
    ),
    (
        "Set 3",
        [np.array([1, 1,  0], dtype=float),
         np.array([0, 1, -1], dtype=float),
         np.array([1, 0,  1], dtype=float)]
    ),
]

for name, vecs in sets:
    independent, rank, M = is_linearly_independent(vecs)
    print(name)
    print("Vectors =", vecs)
    print("Matrix with vectors as columns =\n", M)
    print("rank =", rank, "out of", M.shape[1])
    print("Linearly independent? =", independent)
    print()

print("Explanation:")
print("A set of k vectors is linearly independent iff the matrix formed by stacking them as columns has rank k.")
print("If any set contains the zero vector, it is automatically linearly dependent.")
```

```python
print_header("Question 17: True/False: det(A) is always 0 if A = [c;d][a b]")

def build_outer(a, b, c, d):
    u = np.array([[c], [d]], dtype=float)
    vT = np.array([[a, b]], dtype=float)
    return u @ vT

# Demonstrate with several random choices
rng = np.random.default_rng(42)

for t in range(5):
    a, b, c, d = rng.integers(-5, 6, size=4)
    A = build_outer(a, b, c, d)
    detA = float(np.linalg.det(A))
    rankA = int(np.linalg.matrix_rank(A))
    print(f"trial {t+1}: a={a}, b={b}, c={c}, d={d}")
    print("A =\n", A)
    print("det(A) =", detA)
    print("rank(A) =", rankA)
    print()

print("Explanation:")
print("A = u v^T is an outer product, so rank(A) <= 1.")
print("Any 2x2 matrix with rank < 2 is singular, hence det(A) = 0 (up to numerical precision).")
```

```python
print_header("Question 18: Condition on A so that Ae1 and Ae2 are orthogonal")

e1 = np.array([1, 0], dtype=float)
e2 = np.array([0, 1], dtype=float)

# Example matrix (you can replace this with your own A)
A = np.array([[2.0,  1.0],
              [3.0, -6.0]], dtype=float)

Ae1 = A @ e1  # first column of A
Ae2 = A @ e2  # second column of A
dot = float(Ae1 @ Ae2)

print("A =\n", A)
print("\ne1 =", e1, "e2 =", e2)
print("Ae1 =", Ae1)
print("Ae2 =", Ae2)
print("Ae1 · Ae2 =", dot)
print("Orthogonal? =", np.isclose(dot, 0.0))

print("\nExplanation:")
print("Ae1 is the first column of A, and Ae2 is the second column of A.")
print("Condition for orthogonality is (Ae1)^T (Ae2) = 0, i.e., the two columns of A must have dot product 0.")
print("If A = [[p, q],[r, s]], the condition is p q + r s = 0.")
```

```python
print_header("Question 19: Einstein notation for tr(A^4) and numerical verification")

# Example matrix (can be any square matrix)
A = np.array([[1.2, 0.3, -0.1],
              [0.0, 2.0,  0.5],
              [0.4, 0.0,  1.5]], dtype=float)

A4 = np.linalg.matrix_power(A, 4)
trace_direct = float(np.trace(A4))

# Einstein-style contraction: tr(A^4) = A_{ij} A_{jk} A_{kl} A_{li}
trace_einsum = float(np.einsum("ij,jk,kl,li->", A, A, A, A))

print("A =\n", A)
print("\ntr(A^4) via trace(matrix_power) =", trace_direct)
print("tr(A^4) via Einstein contraction =", trace_einsum)
print("Absolute difference =", abs(trace_direct - trace_einsum))

print("\nExplanation:")
print("Einstein notation for tr(A^4) is A_{ij} A_{jk} A_{k\\ell} A_{\\ell i} (sum over repeated indices).")
print("np.einsum implements exactly this repeated-index summation, matching np.trace(A^4).")
```

```python
print_header("Question 21: Eigenvalues & eigenvectors of A = [[2,1],[0,2]]")

A = np.array([[2, 1],
              [0, 2]], dtype=float)

# General matrix => use eig
eigenvalues, eigenvectors = np.linalg.eig(A)

print("A =\n", A)
print("\nEigenvalues =", eigenvalues)
print("Eigenvectors (columns correspond to eigenvalues) =\n", eigenvectors)

print("\nVerification: residuals A v - lambda v")
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]
    residual = A @ v - lam * v
    print(f"i={i}, lambda={lam:.4f}, residual={residual}")

# Show nullspace direction of (A - 2I) via SVD (numerical)
B = A - 2*np.eye(2)
U, S, VT = np.linalg.svd(B)

print("\nA - 2I =\n", B)
print("Singular values of (A - 2I) =", S)

null_vec = VT.T[:, -1]  # direction with smallest singular value
print("One nullspace direction (eigenvector direction for lambda=2) =", null_vec)
print("Check (A - 2I) @ null_vec =", B @ null_vec)

# Short explanation (printed)
print("\nExplanation:")
print("A is upper-triangular, so its eigenvalues are the diagonal entries: [2, 2].")
print("There is only one independent eigenvector direction (the nullspace of A-2I is 1D),")
print("which is why this matrix is not diagonalizable (it corresponds to a 2x2 Jordan block).")

```

```python
print_header("Question 22: Check if smallest eigenvalue < 0.5 (and Gershgorin bound)")

A = np.array([[3.0, 0.1, 0.3, 1.0],
              [0.1, 1.0, 0.1, 0.2],
              [0.3, 0.1, 5.0, 0.0],
              [1.0, 0.2, 0.0, 1.8]], dtype=float)

print("A =\n", A)

# Symmetric => eigvalsh gives sorted real eigenvalues
eigenvalues = np.linalg.eigvalsh(A)
lam_min = eigenvalues[0]

print("\nEigenvalues (ascending) =", eigenvalues)
print("Smallest eigenvalue =", lam_min)
print("Is smallest eigenvalue < 0.5? =", lam_min < 0.5)

# Gershgorin lower bound
diag = np.diag(A)
radii = np.sum(np.abs(A), axis=1) - np.abs(diag)
lower_bounds = diag - radii
upper_bounds = diag + radii

print("\nGershgorin intervals by row:")
for i in range(A.shape[0]):
    print(f"row {i+1}: center={diag[i]:.4f}, radius={radii[i]:.4f}, interval=[{lower_bounds[i]:.4f}, {upper_bounds[i]:.4f}]")

global_lb = float(np.min(lower_bounds))
print("\nGershgorin global lower bound =", global_lb)
print("So Gershgorin implies lambda_min >= global_lb, hence lambda_min < 0.5 is possible? =", global_lb < 0.5)

# Short explanation (printed)
print("\nExplanation:")
print("np.linalg.eigvalsh computes the exact eigenvalues numerically for a symmetric matrix.")
print("Gershgorin gives a quick bound: every eigenvalue lies in at least one interval shown above.")
print("If the global lower bound is > 0.5, then lambda_min cannot be < 0.5.")

```
